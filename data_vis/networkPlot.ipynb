```python
#This script was written using Python 3.5.2
#This is a script for taking a pandas dataframe of edges and making a network plot using networkx and pygraphviz.
##To run this code you will need you will need networkx (pip install), graphviz (brew or yum install), and pygraphviz (pip install)

%matplotlib notebook
%pylab inline
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx
from networkx.drawing.nx_agraph import graphviz_layout, write_dot, view_pygraphviz
from IPython.display import Image



```

```python
#bring in a dataframe with the columns (Toy dataset): 
##From (starting position), To (ending position), Freq (frequency of occurence in data)
A=pd.DataFrame(data={'From':[1,1,2,3,4,5],'To':[1,5,4,1,1,1], 'Freq':[4,1,1,2,1,3]})
```

```python
#Get values into list format so they can be added as edges
edges=[]
for index, row in A.iterrows():
    x=[]
    x.append(row['From'])
    x.append(row['To'])
    x.append(row['Freq'])
    edges.append(x)

#open a directional graph
G = nx.DiGraph()
```

```python
#Add the edges and weights of the arrows (taken from Freq)
for (u, v, w) in edges:
    G.add_edge(u,v,penwidth=w)

#Draw and save the graph, it will automatically open as well
##prog can be different graphviz display layouts: dot, neato, twopi, circo, or fdp
outFile = open('fileName_dot.png', 'wb')
view_pygraphviz(G,prog='dot',path=outFile)
#this toy dataset with dot layout
Image(filename='fileName_dot.png') 
```

```python
#this toy dataset with neato layout
outFile = open('fileName_neato.png', 'wb')
view_pygraphviz(G,prog='neato',path=outFile)
Image(filename='fileName_neato.png')
```

```python
#this toy dataset with twopi layout
outFile = open('fileName_twopi.png', 'wb')
view_pygraphviz(G,prog='twopi',path=outFile)
Image(filename='fileName_twopi.png')
```

```python
#this toy dataset with circo layout
outFile = open('fileName_circo.png', 'wb')
view_pygraphviz(G,prog='circo',path=outFile)
Image(filename='fileName_circo.png')
```

```python
#this toy dataset with fdp layout
outFile = open('fileName_fdp.png', 'wb')
view_pygraphviz(G,prog='fdp',path=outFile)
Image(filename='fileName_fdp.png')
```
