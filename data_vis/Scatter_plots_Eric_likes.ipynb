```python
#This is a visualization of the function defined in scatter4eric.py
## It builds a scater plot that is seperated and colored based on metadata categories
### and draws the median line through each subset of points
```

```python
%matplotlib notebook
%pylab inline
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Make some toy data to plot
x = np.random.randint(9,size=(50,))
data=pd.DataFrame(data={'Shannon':x,'TimePt':'','Disease State':''})
#Make some metadata to visualize our data by

for i in list(data.index):
    if i < 25:
        data.loc[i,'TimePt']='First'
    if i >= 25:
        data.loc[i,'TimePt']='Second'
    if (data.loc[i,'Shannon'] in [0,1] or data.loc[i,'Shannon'] in [6,7,8]):
        data.loc[i,'Disease State']='Disease'
    if data.loc[i,'Shannon'] in [2,3,4,5]:
        data.loc[i,'Disease State']='Control'

```

```python
#Plot the data points
##this is defined in the function scatter4eric.py

cx=sns.stripplot(x="TimePt", y="Shannon", hue="Disease State", data=data,split=True,jitter=True)
#Get the median line
sns.boxplot(x="TimePt", y="Shannon", hue="Disease State",data=data,
               saturation=0.8,whis=0, fliersize=0,width=0.8,**{'showcaps':False})
plt.setp(cx.artists, alpha=0.5, linewidth=0, fill=False, edgecolor="k")
#build a legen so you don't have a duplicate legend
cx.legend_.remove()
handles, labels = cx.get_legend_handles_labels()
l = plt.legend(handles[0:2], labels[0:2], bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

#A few options to make it prettier 
plt.ylabel('Shannon Diversity')
plt.ylim(-0.1,9)
```

```python
#Change up the visualization so things are seperated by Disease State and colored by time point
#Plot the data points
##this is defined in the function scatter4eric.py

cx=sns.stripplot(x="Disease State", y="Shannon", hue="TimePt", data=data,split=True,jitter=True)
#Get the median line
sns.boxplot(x="Disease State", y="Shannon", hue="TimePt",data=data,
               saturation=0.8,whis=0, fliersize=0,width=0.8,**{'showcaps':False})
plt.setp(cx.artists, alpha=0.5, linewidth=0, fill=False, edgecolor="k")
#build a legen so you don't have a duplicate legend
cx.legend_.remove()
handles, labels = cx.get_legend_handles_labels()
l = plt.legend(handles[0:2], labels[0:2], bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

#A few options to make it prettier 
plt.ylabel('Shannon Diversity')
plt.ylim(-0.1,9)
```
